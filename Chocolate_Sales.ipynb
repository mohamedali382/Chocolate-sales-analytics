# Visualizations

```python
df['month'] = df['Date'].dt.to_period('M')

# Group by month and sum amounts
monthly_amount = df.groupby('month')['Amount'].sum().reset_index()

# Convert Period to datetime for plotting
monthly_amount['month'] = monthly_amount['month'].dt.to_timestamp()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(monthly_amount['month'], monthly_amount['Amount'], marker='o', linestyle='-')
plt.title('Amount Over the Year (Month by Month)')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 6))

sns.lineplot(
    x=df['Date'],
    y=df['Amount'],
    marker='o',    
    color='teal',
    linewidth=1,          # Line width
)

# Titles and labels
plt.title("Amount Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Amount ($)", fontsize=12)

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Add a grid
plt.grid(True, linestyle=':', linewidth=0.7)

# Tight layout for better spacing
plt.tight_layout()

plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# Optional: group by date if there are multiple entries per day
daily_totals = df.groupby('Date')['Amount'].sum().reset_index()

plt.figure(figsize=(20, 6))
sns.histplot(
    data=df,
    x='Date',
    weights='Amount',      # Makes histplot use Amount as weight
    bins=30,               # You can adjust the number of bins (e.g., 12 for months)
    color='teal',
    kde=False
)

# Titles and labels
plt.title("Amount Over Time", fontsize=16, weight='bold')
plt.xlabel("Date", fontsize=13)
plt.ylabel("Total Amount", fontsize=13)

# Rotate x-axis
plt.xticks(rotation=45)

# Add grid
plt.grid(True, linestyle=':', linewidth=0.7)

# Layout
plt.tight_layout()
plt.show()
```

```python
monthly_shippedBoxes = df.groupby('month')['Boxes Shipped'].sum().reset_index()

# Convert Period to datetime for plotting
monthly_shippedBoxes['month'] = monthly_shippedBoxes['month'].dt.to_timestamp()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(monthly_shippedBoxes['month'], monthly_shippedBoxes['Boxes Shipped'], marker='o', linestyle='-')
plt.title('shipped boxes Over the Year (Month by Month)')
plt.xlabel('Month')
plt.ylabel('boxes shipped')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20, 6))

sns.lineplot(
    x=df['Date'],
    y=df['Boxes Shipped'],
    marker='o',    
    color='teal',
    linewidth=1,
)

plt.title("Shipped Boxes Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Box shipped", fontsize=12)

plt.xticks(rotation=45)


plt.grid(True, linestyle=':', linewidth=0.7)

plt.tight_layout()

plt.show()
```

```python
df['month'] = df['Date'].dt.strftime('%b')  # Extract short month name

# Define correct month order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug',]

```

```python
fig, axes = plt.subplots(figsize=(12, 6))

sns.boxplot(x='month', y='Amount', data=df, order=month_order)
plt.title('Amount by Month')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

sns.boxplot(x='month', y='Boxes Shipped', data=df,order=month_order)
plt.title('Boxes Shipped by Month ' \
'for 2022')
plt.grid(True)


plt.tight_layout()
plt.show()
```

```python
country_counts = df['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Frequency']
```

```python
fig = px.scatter_geo(
    country_counts,
    locations="Country",
    locationmode="country names",
    size="Frequency",
    color="Frequency",
    hover_name="Country",
    projection="natural earth",
    title="Frequency of Countries in Dataset"
)

fig.update_layout(
    geo=dict(showland=True, landcolor="lightgray"),
    margin={"r":0,"t":50,"l":0,"b":0}
)

fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=country_counts, x='Frequency', y='Country', palette='gray')

plt.title('Number of Products Paid per country')
plt.xlabel('Frequency')
plt.ylabel('country names')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
country_percents = df['Country'].value_counts(normalize=True) * 100

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_percents, 
        labels=country_percents.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Pastel1.colors)
plt.title('Sales Distribution by Country (%)')
plt.show()
```

```python
df['Weekday'] = df['Date'].dt.day_name() 
```

```python
weekday_counts = df['Weekday'].value_counts().reset_index()
weekday_counts.columns = ['Weekday', 'Count']
```

```python
df.head()
```

```python
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts['Weekday'] = pd.Categorical(weekday_counts['Weekday'], categories=ordered_days, ordered=True)
weekday_counts = weekday_counts.sort_values('Weekday')
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(data=weekday_counts, x='Weekday', y='Count', palette='viridis')

plt.title('Number of Products Paid per Weekday')
plt.xlabel('Weekday')
plt.ylabel('Product Count')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
product_profits = df.groupby('Product')['Amount'].sum().reset_index()
```
